```r
library(data.table)
library(tidyverse)
```

```r
data<-read_csv('Paludisme2015.csv')
```

```r
# handle rows
data<-data[-c(1:4,14,45,60,91,116,132,152,173,184,203),]
data<-data[c(1:189),]
```

```r
# handle cols
data<-data[,c(1,2,seq(4,208,4))]
```

```r
# rename cols
colnames(data)[1:2]<-c("Province", "Region")
for (i in 3:ncol(data)) {
    colnames(data)[i]<-paste("W",i-2)
}
colnames(data)
```

```r
data
```

```r
# write tidy data, end of wrangling
write.csv(data, file = "tidyPaludisme15.csv")
```

```r
# concatinate two years' data
data1<-read_csv('tidyPaludisme14-15.csv')
data2<-read_csv('tidyPaludisme16-17.csv')
data1con2<-cbind(data1[,2:ncol(data1)],data2[,4:ncol(data2)])
data1con2[,54:56]
```

```r
# rename cols
for (i in 3:ncol(data1con2)) {
    colnames(data1con2)[i]<-paste("W",i-2)
}
colnames(data1con2)
```

```r
write.csv(data1con2, file = "tidyPaludisme14-17.csv")
```
