```r
# Convert the numerical data to numerical type in preparation for cross correlation analysis 
cols = c(3:146);    
WHOTotConcat[,cols] = apply(WHOTotConcat[,cols], 2, function(x) as.numeric(as.character(x)))
```

```r
TS1<-as.ts(as.numeric(WHOTotConcat[15, 3:ncol(WHOTotConcat)])) 
TS2<-as.ts(as.numeric(WHOTotConcat[31, 3:ncol(WHOTotConcat)]))
CrossCorr<-ccf(TS1, TS2) 
```

```r
#a <-as.numeric(CrossCorr[-18])
#print(a[1])

#Load the computed scores into a vector 
correlation_vec = c(1:10)
for(u in 1:5)
    {
    score<-as.numeric(CrossCorr[u])
    correlation_vec[u]<-score[1]
     }

for(k in 1:5)
    {
    score<-as.numeric(CrossCorr[-1*k])
    correlation_vec[k+5]<-score[1]
    
    }
#

```
